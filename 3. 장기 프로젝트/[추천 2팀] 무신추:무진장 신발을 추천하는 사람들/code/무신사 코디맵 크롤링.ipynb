# 2. casual

남성 탭을 클릭해야하므로 selenium과 함께 사용

```python
res = requests.get('https://www.musinsa.com/app/codimap/views/19852?style_type=&tag_no=&brand=&display_cnt=60&list_kind=big&sort=date&page=1',  headers={'User-Agent': 'Mozilla/5.0'})
soup = BeautifulSoup(res.content, 'html.parser')

image = soup.find_all('div','codimap-cont')
str(image).split('"')[7]
```

```python
# 크롤링 함수 생성
def crowling_musinsa(pagelink, style_name, df):
    from selenium import webdriver
    from selenium.webdriver.common.by import By
    import time

    driver = webdriver.Chrome('C:\\Korea university\\쿠빅\\22-2 추천시스템 프로젝트\\무신사 크롤링\\chromedriver')
    driver.get(pagelink)
    time.sleep(1)

    x_path = '/html/body/div[3]/div[3]/div[1]/button[2]'
    mencodi = driver.find_element(By.XPATH, x_path)
    mencodi.click()
    html = driver.page_source

    soup = BeautifulSoup(html, 'html.parser')
    content = soup.find_all('a', 'style-list-item__link')

    #코디 제목, 링크 
    title = []
    link = []
    for i in content:
        con = str(i).split('"')
        title.append(con[7])
        li = 'https://www.musinsa.com/app/codimap/views/' + con[5].split("'")[1] + '?style_type=' + style_name + '&tag_no=&brand=&display_cnt=60&list_kind=big&sort=date&page=1'
        link.append(li)

    df['title'] = title
    df['link'] = link

    #코디 상세페이지
    img = []
    txt_total = []
    hashtag_total = []
    for i in df['link']:
        res = requests.get(i,  headers={'User-Agent': 'Mozilla/5.0'})
        soup = BeautifulSoup(res.content, 'html.parser')
        #이미지
        image = soup.find_all('div','codimap-cont')
        image = str(image).split('"')[7]
        #내용
        txt = soup.find_all('p', 'styling_txt')
        txt_final = str(txt).split('>')[1].split('<')[0]
        #해시태그
        hashtag = soup.find_all('div', 'ui-tag-list')
        tag = str(hashtag).split('"')
        hash =[]
        j=7
        while j < len(tag):
            hash.append(tag[j])
            j += 8
        
        img.append(image)
        txt_total.append(txt_final)
        hashtag_total.append(hash)

    df['img'] = img
    df['txt'] = txt_total
    df['hashtag'] = hashtag_total

    driver.close()
```

```python
pagelink = 'https://www.musinsa.com/app/codimap/lists?style_type=casual&tag_no=&brand=&display_cnt=60&list_kind=big&sort=date&page=1'
crowling_musinsa(pagelink, 'casual', casual_m)
casual_m.head()
```
