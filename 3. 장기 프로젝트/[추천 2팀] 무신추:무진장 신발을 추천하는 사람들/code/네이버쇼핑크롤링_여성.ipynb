# 1.단화

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time

chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument('--headless')
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')
driver = webdriver.Chrome('chromedriver', chrome_options = chrome_options)

url = 'https://search.shopping.naver.com/search/category/100000074?catId=50003817%2050003818%2050003821%2050003819%2050003820&mall=204973&origQuery&pagingIndex=1&pagingSize=60&productSet=total&query&sort=review&timestamp=&viewType=list'
driver.get(url)
time.sleep(1)

SCROLL_PAUSE_SEC = 1

last_height = driver.execute_script('return document.body.scrollHeight')

while True:
  driver.execute_script('window.scrollTo(0, document.body.scrollHeight);')
  time.sleep(SCROLL_PAUSE_SEC)
  new_height = driver.execute_script('return document.body.scrollHeight')
  if new_height == last_height:
    break
  last_height = new_height

x_path = '//*[@id="__next"]/div/div[2]/div/div[3]/div[1]/div[1]/div/div[1]/a[4]' #리뷰많은순 xpath
women = driver.find_element(By.XPATH, x_path)
women.click()

html = driver.page_source

soup1 = BeautifulSoup(html, 'html.parser')
```

```python
brand1 = soup1.find_all('a', 'basicList_link__JLQJf')
```

```python
brand = []
for i in brand1:
  con = str(i).split('"')
  con_0 = con[13]
  con_1 = con_0.split(maxsplit=1)[0]
  con_1 = con_1.replace("[", '')
  con_1 = con_1.replace("]", '')
  brand.append(con_1)

flat_w['brand'] = brand
```

```python
img1 = soup1.find_all('a', 'thumbnail_thumb__Bxb6Z')
```

```python
img = []
for i in img1:
  con = str(i).split('"')
  img.append(con[7])

flat_w['img'] = img
```

```python
category1 = soup1.find_all('div', 'basicList_depth__SbZWF')
```

```python
category = []
for i in range(len(category1)):
  category2 = category1[i].get_text()
  # con = str(i).split('"')
  category.append(category2)

flat_w['category'] = category
```

```python
price1 = soup1.find_all('span', 'price_num__S2p_v')
price = []
for i in price1:
  con = str(i).split('>')[1].split('<')[0]
  price.append(con)

flat_w['price'] = price
```

```python
# img url에 하나씩 들어가서 상품 사진이랑 브랜드 이름 가져오기
product_url1 = list(flat_w['img'])
```

```python
market_url = []
for i in range(len(product_url1)):
  wd = webdriver.Chrome('chromedriver', chrome_options = chrome_options)
  wd.get(product_url1[i])
  time.sleep(0.5)
  html = wd.page_source
  soup2 = BeautifulSoup(html, 'html.parser')
  script = soup2.find('script')
  for j in script:
    con = str(j).split('"')
    market_url.append(con[3])

flat_w['brand_url'] = market_url
```

```python
# 각 brand_url에서 상품명, 브랜드명, 이미지 가져오기
# product, brand, product_img 채우기
```

```python
market_urls = list(flat_w['brand_url'])
```

```python
product = []
image = []

for i in range(len(market_urls)):
  wd = webdriver.Chrome('chromedriver', chrome_options = chrome_options)
  wd.get(market_urls[i])
  time.sleep(1)
  html = wd.page_source
  soup3 = BeautifulSoup(html, 'html.parser')

  # 상품명
  product_name = soup3.find_all('span', 'product_title')[0].find('em').get_text()
  product.append(product_name)

  # 이미지
  product_img = soup3.find('img', 'plus_cursor')['src']
  image.append(product_img)

flat_w['product'] = product
flat_w['product_img'] = image
```

# 2. sneakers

```python
# 크롤링 함수 생성
def crowling_naver(pagelink, df):
    from selenium import webdriver
    from selenium.webdriver.common.by import By
    from selenium.webdriver.common.keys import Keys
    import time

    chrome_options = webdriver.ChromeOptions()
    chrome_options.add_argument('--headless')
    chrome_options.add_argument('--no-sandbox')
    chrome_options.add_argument('--disable-dev-shm-usage')
    driver = webdriver.Chrome('chromedriver', chrome_options = chrome_options)

    url = pagelink
    driver.get(url)
    time.sleep(1)

    SCROLL_PAUSE_SEC = 1

    last_height = driver.execute_script('return document.body.scrollHeight')

    while True:
      driver.execute_script('window.scrollTo(0, document.body.scrollHeight);')
      time.sleep(SCROLL_PAUSE_SEC)
      new_height = driver.execute_script('return document.body.scrollHeight')
      if new_height == last_height:
        break
      last_height = new_height

    x_path = '//*[@id="__next"]/div/div[2]/div/div[3]/div[1]/div[1]/div/div[1]/a[4]'
    women = driver.find_element(By.XPATH, x_path)
    women.click()
    html = driver.page_source

    soup = BeautifulSoup(html, 'html.parser')

    # 브랜드
    brand1 = soup.find_all('a', 'basicList_link__JLQJf')
    brand = []
    for i in brand1:
      con = str(i).split('"')
      con_0 = con[13]
      con_1 = con_0.split(maxsplit=1)[0]
      con_1 = con_1.replace("[", '')
      con_1 = con_1.replace("]", '')
      brand.append(con_1)

    df['brand'] = brand

    # 상품 url 불러오기
    img1 = soup.find_all('a', 'thumbnail_thumb__Bxb6Z')
    img = []
    for i in img1:
      con = str(i).split('"')
      img.append(con[7])

    df['img'] = img

    #카테고리
    category1 = soup.find_all('div', 'basicList_depth__SbZWF')
    category = []
    for i in range(len(category1)):
      category2 = category1[i].get_text()
      category.append(category2)

    df['category'] = category

    #가격
    price1 = soup.find_all('span', 'price_num__S2p_v')
    price = []
    for i in price1:
      con = str(i).split('>')[1].split('<')[0]
      price.append(con)

    df['price'] = price
        
    # img url에 하나씩 들어가서 상품 사진이랑 브랜드 이름 가져오기
    product_url1 = list(df['img'])

    market_url = []
    for i in range(len(product_url1)):
      wd = webdriver.Chrome('chromedriver', chrome_options = chrome_options)
      wd.get(product_url1[i])
      time.sleep(0.5)
      html1 = wd.page_source
      soup2 = BeautifulSoup(html1, 'html.parser')
      script = soup2.find('script')
      for j in script:
        con = str(j).split('"')
        market_url.append(con[3])

    df['brand_url'] = market_url

    # 각 brand_url에서 상품명, 브랜드명, 이미지 가져오기
    # product, brand, product_img 채우기
    market_urls = list(df['brand_url'])

    product = []
    image = []

    for i in range(len(market_urls)):
      wd = webdriver.Chrome('chromedriver', chrome_options = chrome_options)
      wd.get(market_urls[i])
      time.sleep(1)
      html3 = wd.page_source
      soup3 = BeautifulSoup(html3, 'html.parser')

      # 상품명
      product_name = soup3.find_all('span', 'product_title')[0].find('em').get_text()
      product.append(product_name)

      # 이미지
      product_img = soup3.find('img', 'plus_cursor')['src']
      image.append(product_img)

    df['product'] = product
    df['product_img'] = image
```

```python
pagelink = 'https://search.shopping.naver.com/search/category/100000075?catId=50001463%2050003822&mall=204973&origQuery&pagingIndex=1&pagingSize=60&productSet=total&query&sort=review&timestamp=&viewType=list'
crowling_naver(pagelink, sneakers_w)
```
