```python
import pandas as pd
import numpy as np
import xgboost as xgb
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error 
from sklearn.ensemble import RandomForestRegressor, VotingRegressor
from sklearn.linear_model import Ridge, Lasso, LinearRegression
from lightgbm import LGBMRegressor
from sklearn.naive_bayes import BernoulliNB
from sklearn.tree import DecisionTreeRegressor
```

```python
#Import train & test
test_label = pd.read_csv('/content/drive/MyDrive/(2차) 22-2 KUBIG 분류 예측 1팀/노연수/encoding data/test_label.csv')
train_label = pd.read_csv('/content/drive/MyDrive/(2차) 22-2 KUBIG 분류 예측 1팀/노연수/encoding data/train_label.csv')
test_none = pd.read_csv('/content/drive/MyDrive/(2차) 22-2 KUBIG 분류 예측 1팀/노연수/encoding data/test_none.csv')
train_none = pd.read_csv('/content/drive/MyDrive/(2차) 22-2 KUBIG 분류 예측 1팀/노연수/encoding data/train_none.csv')
test_onehot = pd.read_csv('/content/drive/MyDrive/(2차) 22-2 KUBIG 분류 예측 1팀/노연수/encoding data/test_onehot.csv')
train_onehot = pd.read_csv('/content/drive/MyDrive/(2차) 22-2 KUBIG 분류 예측 1팀/노연수/encoding data/train_onehot.csv')
```

```python
X_label = train_label.drop(['rating', 'nickname'],axis=1)
y_label = train_label['rating']
X_label_train,X_label_test,y_label_train,y_label_test = train_test_split(X_label,y_label,test_size=0.2,random_state=522)

X_one = train_onehot.drop(['rating', 'nickname'],axis=1)
y_one = train_onehot['rating']
X_one_train,X_one_test,y_one_train,y_one_test = train_test_split(X_one,y_one,test_size=0.2,random_state=522)

train_none = train_none[train_none.T[train_none.dtypes!=np.object].index]
X_none = train_none.drop(['rating'],axis=1)
y_none = train_none['rating']
X_none_train,X_none_test,y_none_train,y_none_test = train_test_split(X_none,y_none,test_size=0.2,random_state=522)
```

```python
#XGBregressor
my_model = xgb.XGBRegressor(learning_rate=0.1,max_depth=5,n_estimators=100)
my_model.fit(X_label_train, y_label_train, verbose=False)
label_pred = my_model.predict(X_label_test) 

my_model.fit(X_one_train, y_one_train, verbose=False)
one_pred = my_model.predict(X_one_test) 

my_model.fit(X_none_train, y_none_train, verbose=False)
none_pred = my_model.predict(X_none_test) 
```

```python
label_RMSE = mean_squared_error(y_label_test, label_pred)**0.5
one_RMSE = mean_squared_error(y_one_test, one_pred)**0.5
none_RMSE = mean_squared_error(y_none_test, none_pred)**0.5
print(label_RMSE)
print(one_RMSE)
print(none_RMSE)
```

```python
#RandomForestRegressor
my_model = RandomForestRegressor(random_state=0, max_depth=5, min_samples_leaf=8, min_samples_split=8,n_estimators=100)
my_model.fit(X_label_train, y_label_train)
label_pred = my_model.predict(X_label_test)

my_model.fit(X_one_train, y_one_train)
one_pred = my_model.predict(X_one_test) 

my_model.fit(X_none_train, y_none_train)
none_pred = my_model.predict(X_none_test)
```

```python
label_RMSE = mean_squared_error(y_label_test, label_pred)**0.5
one_RMSE = mean_squared_error(y_one_test, one_pred)**0.5
none_RMSE = mean_squared_error(y_none_test, none_pred)**0.5
print(label_RMSE)
print(one_RMSE)
print(none_RMSE)
```

```python
#Ridge
my_model = Ridge(alpha=1)
my_model.fit(X_label_train, y_label_train)
label_pred = my_model.predict(X_label_test)

my_model.fit(X_one_train, y_one_train)
one_pred = my_model.predict(X_one_test) 

my_model.fit(X_none_train, y_none_train)
none_pred = my_model.predict(X_none_test)
```

```python
#voting
#onehot이 모든 모델에서 값이 잘나옴 / 제일 잘나온 3개 모델 사용
xxgb = xgb.XGBRegressor(learning_rate=0.1,max_depth=5,n_estimators=100)
lgbm = RandomForestRegressor(random_state=0, max_depth=5, min_samples_leaf=8, min_samples_split=8,n_estimators=100)
randomForest = LGBMRegressor()

models = [
    ('xgb', xxgb), 
    ('lgbm', lgbm), 
    ('randomForest', randomForest), 
]
voting_regressor = VotingRegressor(models, n_jobs=-1)
voting_regressor.fit(X_one_train, y_one_train)
voting_pred = voting_regressor.predict(X_one_test)
```

```python
voting_RMSE = mean_squared_error(y_one_test, voting_pred)**0.5
print(voting_RMSE)
```
